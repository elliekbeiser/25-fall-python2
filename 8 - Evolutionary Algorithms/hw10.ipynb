````python
import random

def value(item_indices):
    global items
    total_value = 0
    for item_index in item_indices:
        total_value += items[item_index][1]  # value is at index 1
    return total_value

def weight(item_indices):
    global items
    total_weight = 0
    for item_index in item_indices:
        total_weight += items[item_index][0]  # weight is at index 0
    return total_weight

def knapsack(capacity, optimal_value=None):
    global items

    if len(items) == 0:
        return [], 0
    
    max_initial_items = min(len(items), capacity)
    initial_size = random.randint(1, max_initial_items) if max_initial_items > 0 else 1
    
    knapsack_items = random.sample(range(len(items)), initial_size)
    most_value = value(knapsack_items)
    
    for generation in range(1, 101):
        parent = knapsack_items.copy()
        
        for _ in range(100):
            offspring = parent.copy()
            offspring_weight = weight(offspring)
            
            # If overweight, remove a random item
            if offspring_weight > capacity and len(offspring) > 0:
                random_index = random.randint(0, len(offspring) - 1)
                offspring.pop(random_index)
                offspring_weight = weight(offspring)
            
            # Add a random item if capacity allows
            available_items = [i for i in range(len(items)) if i not in offspring]
            if available_items:
                new_item = random.choice(available_items)
                if offspring_weight + items[new_item][0] <= capacity:  # weight is at index 0
                    offspring.append(new_item)
                    offspring_weight = weight(offspring)
            
            offspring_value = value(offspring)
            offspring_weight = weight(offspring)
        
            if offspring_value > most_value and offspring_weight <= capacity:
                knapsack_items = offspring.copy()
                most_value = offspring_value
        
        print(f"Generation {generation}: {most_value}")
        
        if optimal_value is not None and most_value >= optimal_value:
            break
    
    return #knapsack_items, most_value
````
