```python
# Just as def defines a function but does not call it, 
# class defines an object but does not instantiate one
class Point:
    '''
    Represents a point in 2D space

    attributes: x, y
    '''

    # Constructor always requires self as the first argument, followed by whatever attributes you need
    # The attributes here have been assigned default values
    def __init__(self, input_x = 0, input_y = 0):
        self.x = input_x
        self.y = input_y
```

```python
# One purpose of classes is to bundle together (encapsulate) related data so it can be
# reused in other classes/functions.
class Rectangle:
    '''
    Represents a rectangle with lower left corner at corner.

    attributes: corner (Point object), height (float), width (float)
    '''

    # PRACTICE: Define the constructor below
    def __init__(self, input_corner = Point(), input_height = 0, input_width = 0):
        self.corner = input_corner
        self.height = input_height
        self.width = input_width
```

```python
def get_corners(rectangle = Rectangle()):
    '''
    takes rectangle object as input
    output: a list of point objects: one for each corner of rectangle
    '''
    return rectangle.corner.x, rectangle.corner.y
    #corners = [corner1, corner2, corner3, corner4]
    #return corners
```

```python
rectangle = Rectangle(Point(1,2),3,4)
get_corners(rectangle)
```

```python
def distance(inputPoint = Point(), inputPoint2 = Point()):
    '''
    takes two point objects as inputs and returns the distance between them
    import math to use square root function from python built-ins
    '''
    import math

    x_coord = inputPoint.x
    y_coord = inputPoint.y
    x_coord2 = inputPoint2.x
    y_coord2 = inputPoint2.y

    totalDist = math.sqrt((x_coord2 - x_coord) ** 2 + (y_coord2 - y_coord) ** 2)
    return totalDist
```
