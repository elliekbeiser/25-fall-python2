```python
class Point:
    '''
    Represents a point in 2D space

    attributes: x, y
    '''

    def __init__(self, input_x = 0, input_y = 0):
        self.x = input_x
        self.y = input_y

    def __str__(self):
        return f'({self.x}, {self.y})'
```

```python
class Circle:
    '''
    attributes: center as a point object and radius as a float
    '''
    def __init__(self, center = Point(), radius = 0):
        self.center = center
        self.radius = radius

    def __str__(self):
        return f'Center: {self.center}, Radius: {self.radius}'
    
    def get_area(self):
        from math import pi
        area = pi * (self.radius * self.radius)
        return area
    
    def to_square(self):
        import math
        side = math.sqrt(self.get_area())
        corner_x = self.center.x - side/2
        corner_y = self.center.y - side/2
        return f'corner: {corner_x,corner_y}, height and width: {side}'
```

```python
circ = Circle(Point(), 1)
print(circ)
print(circ.get_area())
print(circ.to_square())
```

```python
class Rectangle:
    '''
    Represents a rectangle with lower left corner at corner.

    attributes: corner (Point object), height (float), width (float)
    '''

    def __init__(self, input_corner = Point(), input_height = 0, input_width = 0):
        self.corner = input_corner
        self.height = input_height
        self.width = input_width
    
    def get_area(self):
        area = self.height * self.width
        return area
    
    def __lt__(self, rectangle2):
        if self.get_area() < rectangle2.get_area():
            return True
        else:
            return False
        
    def get_center_point(self):
    
        input_x = self.corner.x
        input_y = self.corner.y
        height = self.height
        width = self.width

        midpoint_x = input_x + width / 2
        midpoint_y = input_y + height / 2

        midpoint = Point(midpoint_x, midpoint_y)
        return midpoint
    
    def to_circle(self):
        import math
        radius = math.sqrt(self.get_area()/math.pi)
        return Circle(self.get_center_point(), radius)
```

```python
rect = Rectangle(Point(),2,2)
print(rect.to_circle())
```
